# 🚀 Missão Espacial para Marte
Sistema de Verificação de Decolagem


## 1.1 Organização da Telemetria

A telemetria da nave espacial consiste no monitoramento de variáveis críticas para garantir a segurança da missão. Os principais parâmetros analisados são:

- **Temperatura interna e externa:** fundamentais para garantir o funcionamento adequado dos sistemas e a integridade da tripulação.
- **Integridade estrutural:** representada por valores binários (0 = comprometida, 1 = íntegra).
- **Níveis de energia:** indicam a porcentagem de carga disponível para a missão.
- **Pressão dos tanques:** essencial para o correto funcionamento dos sistemas de propulsão.
- **Status dos módulos críticos:** verifica se sistemas essenciais estão operacionais.

Esses dados são utilizados para determinar se a nave está apta para decolagem.


## 1.2 Algoritmo de Verificação

```text
INÍCIO

Ler temperatura interna
Ler temperatura externa
Ler integridade estrutural
Ler nível de energia
Ler pressão dos tanques
Ler status dos módulos críticos

SE (temperatura interna entre 18 e 27) E
   (temperatura externa entre -50 e 50) E
   (integridade estrutural == 1) E
   (energia >= 70) E
   (pressão dos tanques >= 80) E
   (status dos módulos == OK)

    Exibir "PRONTO PARA DECOLAR"

SENÃO
    Exibir "DECOLAGEM ABORTADA"

FIM
```


````python
# Simulação de telemetria - Missão Marte

temperatura_interna = 22
temperatura_externa = -20
integridade_estrutural = 1
nivel_energia = 85
pressao_tanques = 90
modulos_criticos = True

print("=== VERIFICAÇÃO DE DECOLAGEM ===")

if (18 <= temperatura_interna <= 27 and
    -50 <= temperatura_externa <= 50 and
    integridade_estrutural == 1 and
    nivel_energia >= 70 and
    pressao_tanques >= 80 and
    modulos_criticos):

    print("PRONTO PARA DECOLAR 🚀")

else:
    print("DECOLAGEM ABORTADA ❌")
````
